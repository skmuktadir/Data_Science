# Linear Regression with Gradient Decent 

```python
# Step 1
m = 0
c = 0
n = len(x)
learning_rate = 0.0001

# Step 2
for i in range(500):
    predicted_y = (m * x) + c
    Dm = (-2 / n) * sum(x * (y - predicted_y))
    # Step 3
    Dc = (-2 / n) * sum(y - predicted_y)
    # Update the weights in each iteration
    m = m - (learning_rate * Dm)
    c = c - (learning_rate * Dc)

print(m, c)
```

```python
sum_OLS = []
sum_OLS_sq= []
sum_yi_Y__OLS_sq=0
sum_yi_Y__OLS = 0
predicted_y__OLS = []
dif_y_predicted_y__OLS = []
count = 0

for i in x:
    # Append a placeholder value for predicted_y__OLS
    y=df['y']
    predicted_y__OLS.append(None)
    # Assuming reg is your regression model, you can't assign a value to an index that doesn't exist in the list yet, so we need to use append instead.
    predicted_y__OLS[count] = reg.predict([[i]])  # OLS MODEL predicted y
    print("When x =", i, "Actual y=",y[count], "then Predicted_y_OLS =", predicted_y__OLS[count], " And Predicted_y =", (m * i) + c)
    dif_y_predicted_y__OLS.append(y[count] - predicted_y__OLS[count])  # difference between actual y vs OLS MODEL predicted y
    #print("dif_y_predicted_y__OLS[", count, "] =", 
    dif_y_predicted_y__OLS[count]
    sum_OLS_sq.append((dif_y_predicted_y__OLS[count])**2)
    sum_OLS.append(dif_y_predicted_y__OLS[count])
    sum_yi_Y__OLS_sq=sum_yi_Y__OLS_sq+sum_OLS_sq[count]
    sum_yi_Y__OLS = sum_yi_Y__OLS + sum_OLS[count]  # sum of the difference between actual y vs OLS MODEL predicted y
    #print('sum_yi_Y__OLS : [',count,']',sum_yi_Y__OLS)
    count = count + 1

print("Total sum : ",(sum_yi_Y__OLS))

```
