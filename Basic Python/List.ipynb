# List in a list

```python
list2=['Apple',['Hello',2,3,4],'Banana','Orange']
print (list2)
```

```python
len (list2)
```

```python
for i in range (len(list2)):
    if isinstance(list2[i], list): # isinstance(obj, class_or_tuple, /)
         for j in range (len(list2)):
              print (list2[i][j],sep='',end='\n')
    else:
        print(list2[i])
```
